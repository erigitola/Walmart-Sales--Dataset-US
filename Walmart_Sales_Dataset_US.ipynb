```python
gender_counts = df['Gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue','orange'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

```

```python
import seaborn as sns
```

```python
sns.boxplot(x='Gender', y='Purchase', data=df)
plt.title('Average Purchase by Gender')
plt.show()
```

```python
sns.boxplot(x='City_Category', y='Purchase',data=df)
plt.title('Average Purchase by City Category')
plt.show()
```

```python
sns.boxplot(x='City_Category', y='Purchase',data=df, hue='Gender')
plt.title('Purchase by Category')
plt.show()
```

```python
sns.violinplot(x='City_Category', y='Purchase', data=df, hue= 'Gender', split=True)
plt.title('Purchase by Category')
plt.show()
```

```python
sns.barplot(x='Product_Category', y= 'Purchase', hue='City_Category', data=df)
plt.title('Average Purchase by Product and City')
plt.xlabel('Product Catgory')
plt.ylabel('Average Purchase')
plt.legend(title='City')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x='City_Category', data=df, palette='pastel')
plt.title('Customer Distribution by City')
plt.xlabel('City_Category')
plt.ylabel('Count')
plt.show()
```

```python
occupation_counts = df['Occupation'].value_counts()
plt.pie(occupation_counts,labels=occupation_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Occupation Distribution')
plt.axis('equal')
plt.show()
```

```python
cross_tab = pd.crosstab(df['City_Category'], df['Stay_In_Current_City_Years'])

sns.heatmap(cross_tab, annot=True, cmap='coolwarm', fmt='d')
plt.title('Customer Count by City and Stay Duration')
plt.xlabel('Stay in Current City (Years)')
plt.ylabel('City Category')
plt.show()
```

```python
sns.catplot(x='Age', hue='Marital_Status', col='Gender',data=df, kind='count', palette='Set1')
plt.subplots_adjust(top=0.85)
plt.suptitle('Age vs Marital Status Split by Gender')
plt.show()
```

```python
ct = pd.crosstab(index=[df['Gender'], df['City_Category']], columns=df['Stay_In_Current_City_Years'])

sns.heatmap(ct, annot=True,cmap='Blues')
plt.title('Stay Duration by Gender and City')
plt.xlabel('Stay Duration')
plt.ylabel('Stay (Years)')
plt.ylabel('Gender+ City')
plt.show()
```

```python
sns.countplot(x='City_Category', hue='Stay_In_Current_City_Years',data=df, palette='Set3')
plt.title('Stay Duration by City Category')
plt.xlabel('City Category')
plt.ylabel('Count')
plt.legend(title='Stay In Years')
plt.show()
```

```python
sns.catplot(
    x='Product_Category',
    y='Purchase',
    hue='Gender',
    col='City_Category',
    row='Stay_In_Current_City_Years',
    kind='bar',
    data=df,
    palette='Set2',
    height=4,
    aspect=1
)
```

```python
sns.boxplot(x='Product_Category', y='Purchase', hue='Gender', data=df[df['City_Category'] == 'A'])
plt.title('Purchase Distribution by Product & Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group and Pivot
pivot = df.groupby(['Gender', 'City_Category', 'Stay_In_Current_City_Years', 'Product_Category'])['Purchase'].mean().unstack()

# Show Just one slice(e.g, Gender='Male')

heatmap_data = pivot.loc['M'] # Filter by other levels

# Heatmap
sns.heatmap(heatmap_data, annot=True, fmt='.1f', cmap='coolwarm')
plt.title('Average Purchase (Male) by City, Stay, & Product')
plt.show()
```
